```python
# read and removal of all empty value
from csv import reader

groceries = []
with open(r'Dataset_Day15.csv') as csvfile:
    read = reader(csvfile)
    for row in read:
        # Remove 'NaN' and empty strings
        row_new = [x for x in row if x != 'NaN' and x.strip() != '']
        groceries.append(row_new)
```

```python
#Transform the data using TransactonEncoder to perform Market Basket Analysis.
from mlxtend.preprocessing import TransactionEncoder

encoder = TransactionEncoder()

transactions = encoder.fit(groceries).transform(groceries)

import pandas as pd

itemsets = pd.DataFrame(transactions, columns = encoder.columns_)
display(itemsets.head())
display(itemsets.info())
```

```python
#Use min_support = 0.02 to find frequent itemsets.
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(itemsets, min_support = 0.02, use_colnames = True)
frequent_itemsets
```

```python
frequent_itemsets['length'] = frequent_itemsets['itemsets'].str.len()
frequent_itemsets.head(-5)
```

Observations:-

Frequent Single Items: Items like burgers (8.7%), spaghetti (not shown in full list but likely high), and brownies (3.37%) are among the most frequent individual products. Even relatively less common items like almonds (2.03%) and butter (3.01%) meet the minimum support of 0.02, thus getting included.
Frequent Pairs (Length = 2): Common pairs include “spaghetti & mineral water” (5.97%), “pancakes & mineral water” (3.37%), and “shrimp & mineral water” (2.36%).
Mineral Water: It appears to be a key product it several purchases as a highly associated item.
Itemset Size: In this table, we only see itemsets of sizes 1 and 2, it is likely that any larger itemsets are way less common so it falls below the frequency threshold.
Strategy: From this information, we can improve the way the supermarkets place their products in their shelves, strategically placing the common pairs of items close to each other.

```python
#Find top 5 antecedent -> consequent rules based on ‘conviction’, ‘leverage’, ‘lift’. Explain your findings.
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.15)
rules
```

Observations:-

Olive Oil → Spaghetti has highest lift (about 1.999) and conviction (about 1.267), meaning olive oil is a great predictor of spaghetti purchases.
Rules like Burgers → Eggs and Burgers → French Fries have conviction above 1.22 and lift above 1.83, this means burgers can strongly predict the purchases for Eggs and French Fries.
Rules like Shrimp → Spaghetti or Pancakes → Spaghetti have lower leverages each, but they show significant lift. This suggests that while these combinations are unconventional, they are supported by consistent customers.

```python
# Top 5 rules by conviction
top5_conviction = rules.sort_values('conviction', ascending=False).head(5)
# Top 5 rules by leverage
top5_leverage = rules.sort_values('leverage', ascending=False).head(5)
# Top 5 rules by lift
top5_lift = rules.sort_values('lift', ascending=False).head(5)
top5_conviction
```

Observations:-

Spaghetti → Ground Meat is the strongest rule in the list, having a lift of 2.29.
Soup → Mineral Water is the strongest rule by conviction which makes sense because without the purchase of mineral water, it is less likely that soup will be purchased.
Rules like ground meat → spaghetti and olive oil → spaghetti can be used for targeted marketing, maybe in the form of combos or dedicated sections, etc..

```python
#Find top 2 & bottom 2 antecedent -> consequent rules based on ‘zhang’s metric’. Explain your findings.
display(top5_conviction[['antecedents', 'consequents', 'conviction']])
display(top5_leverage[['antecedents', 'consequents', 'leverage']])
display(top5_lift[['antecedents', 'consequents', 'lift']])
```

```python
display(rules.sort_values('zhangs_metric',ascending=False).head(2))
display(rules.sort_values('zhangs_metric',ascending=True).head(2))
```

Observations:-

Spaghetti → Ground Meat and Ground Meat → Spaghetti show the highest Zhang Metric. This suggests that when one of these items is bought, the other's purchase can be reliably predicted.
French Fries → Mineral Water and Spaghetti → French Fries show Zhang Metric less than 0. These rules show slightly negative to weak association. So it is unlikely that if a customer purchases French Fries, they will also get Mineral Water.
