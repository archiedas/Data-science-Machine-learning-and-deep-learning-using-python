```python
#Histogram
ColumnsToVisualize = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]

df[ColumnsToVisualize].hist(figsize=(20, 10))
```

Observations:

RM, PTRATIO, AGE, NOX are closer to normal.

```python
#Boxplot
sns.boxplot(data=df[ColumnsToVisualize],orient="v",palette = "Set2")
```

Observations:

TAX has a very high range and many outliers.

```python
#HeatMap
sns.heatmap(df.select_dtypes(include='number').corr())
```

Observations:

MEDV is strongly negatively correlated with LSTAT
MEDV is strongly positively correlated with RM
TAX, RAD, INDUS, NOX are positively correlated among themselves.
DIS is negatively correlated with NOX, TAX, RAD, INDUS, AGE

```python
#2.	Create a simple linear regression model that quantitatively relates ‘MEDV’ with ‘RM’. (10 marks)
    #a.	Share the model performance metrics and print the full regression model with coefficients.
    #b.	Use the model to predict the price of the house for ‘RM’ = 7
from sklearn.model_selection import train_test_split 

X = pd.DataFrame(df["RM"])
y = pd.DataFrame(df["MEDV"])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm = lm.fit(X_train, y_train)
display(lm.coef_) 
display(lm.intercept_) 
```

```python
predict_price = lm.predict(np.array([[7]]))
print("Predicted price: ", predict_price[0])
```
